# Univariate Analysis

```python
# 1) Categorical Columns
cat_cols = ['Category', 'Sub-Category', 'Region', 'Segment', 'Ship Mode']

for col in cat_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df, palette="coolwarm", edgecolor='black')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 2) Numerical Columns
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit']

df[num_cols].hist(figsize=(10, 6), bins=30, color='#1f77b4', edgecolor='black')
plt.suptitle("Distribution of Numerical Columns", fontsize=14)
plt.show()
```

# Bivariate Analysis

```python
# a) Sales vs Profit (Relationship)
plt.figure(figsize=(7,5))
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category', palette='Set2')
plt.title("Sales vs Profit by Category")
plt.show()
```

```python
# b) Category-wise Profit
plt.figure(figsize=(7,4))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum, palette='Set3', edgecolor='black')
plt.title("Total Profit by Category")
plt.show()
```

```python
# C) Sub-Category vs Sales
plt.figure(figsize=(10,4))
sns.barplot(x='Sub-Category', y='Sales', data=df, estimator=sum, palette='crest', edgecolor='black')
plt.title("Sales by Sub-Category")
plt.xticks(rotation=60)
plt.show()
```

```python
# D) Region vs Profit\
plt.figure(figsize=(6,4))
sns.barplot(x='Region', y='Profit', data=df, estimator=sum, palette='magma')
plt.title("Profit by Region")
plt.show()
```

# Outlier Detection

```python
# Using Boxplots
for col in ['Sales', 'Profit', 'Discount']:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

# Correlation Analysis

```python
plt.figure(figsize=(6,4))
corr = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='RdBu_r', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

# Time-based Trend Analysis

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Yearly Sales
plt.figure(figsize=(6,4))
sns.barplot(x='Year', y='Sales', data=df, estimator=sum, palette='viridis')
plt.title("Yearly Sales Trend")
plt.show()

# Monthly Sales Trend
monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(8,4))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='orange')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Top & Bottom Performers

```python
# a) Top 10 Profitable States
top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,4))
sns.barplot(x=top_states.values, y=top_states.index, palette='Greens_r')
plt.title("Top 10 Profitable States")
plt.xlabel("Profit")
plt.show()
```

```python
# b) Bottom 10 Loss-Making States
loss_states = df.groupby('State')['Profit'].sum().sort_values().head(10)
plt.figure(figsize=(8,4))
sns.barplot(x=loss_states.values, y=loss_states.index, palette='Reds_r')
plt.title("Bottom 10 Loss-Making States")
plt.xlabel("Profit")
plt.show()
```

# Segment & Ship Mode Analysis

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Segment', y='Profit', data=df, estimator=sum, palette='ch:s=-.2,r=.6')
plt.title("Profit by Segment")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='Ship Mode', y='Sales', data=df, estimator=sum, palette='Blues')
plt.title("Sales by Ship Mode")
plt.show()
```
