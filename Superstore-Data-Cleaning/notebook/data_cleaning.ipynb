```python
df.isnull().sum() 
#this will output as where's the missing value exist.
#if each columns got 0 means there no missing value else these are the number of missing value exist
```

```python
df.drop_duplicates(inplace=True)
#simply removes the duplicates
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
#this will convert Order Daete and Ship Date from sting to actual Datatime objects
```

```python
df.info() #Now, u can see the Changes
```

```python
#convert postal code from object to string
df['Postal Code']=df['Postal Code'].astype(str)
```

```python
df['Sales']=df['Sales'].astype(float)
df['Profit']=df['Profit'].astype(float)
df['Discount']=df['Discount'].astype(float)
df['Quantity']=df['Quantity'].astype(int)
#Ensure Numeric Columns are numeric
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
#Extract year, month from Order Date
```

```python
df['Ship_Delay'] = (df['Ship Date'] - df['Order Date']).dt.days
#Shipping Delays
```

```python
df['Profit_Margin'] = df['Profit'] / df['Sales']
#Profit Margins
```

```python
df['Category'] = df['Category'].str.strip().str.title()
df['City'] = df['City'].str.strip().str.title()
df['State'] = df['State'].str.strip().str.title()
df['Segment'] = df['Segment'].str.strip().str.title()
#these will removes the Extra Spces and Capatalizes text in Title cases
#str.strip() removes the Extra Spaces
#str.title() Capatalizes the Title Cases
```

```python
df.drop(['Row ID', 'Country'], axis=1, inplace=True)
#These Columns are not provinding any neccesary values so let drop them
#Row ID and Country(always United States)
```
